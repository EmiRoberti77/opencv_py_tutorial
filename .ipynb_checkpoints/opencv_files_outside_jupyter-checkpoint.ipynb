```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
blank_img = np.zeros(shape=(512, 512,3), dtype=np.int16)
```

```python
blank_img.shape
```

```python
plt.imshow(blank_img)
```

```python
plt.show()
```

```python
cv2.rectangle(blank_img,pt1=(340,10), pt2=(500,100),color=(0,255,0), thickness=10)
```

```python
plt.imshow(blank_img)
plt.show()
```

```python
cv2.rectangle(blank_img, pt1=(200,200), pt2=(300,300), color=(0,0,255), thickness=10)
plt.imshow(blank_img)
plt.show()
```

```python
cv2.circle(img=blank_img, center=(100,100), radius=50, color=(255,0,0), thickness=8)
plt.imshow(blank_img)
plt.show()
```

```python
cv2.circle(img=blank_img, center=(400,400), radius=50, color=(255,0,0), thickness=-1)
plt.imshow(blank_img)
plt.show()
```

```python
cv2.line(blank_img, pt1=(0,0), pt2=(512,512), color=(255,100,0), thickness=5)
plt.imshow(blank_img)
plt.show()
```

```python
font = cv2.FONT_HERSHEY_SIMPLEX
cv2.putText(blank_img, 
            text='Emi', 
            org=(10,500), 
            fontFace=font, 
            fontScale=4, 
            color=(255,255,255), 
            thickness=3, 
            lineType=cv2.LINE_AA)
plt.imshow(blank_img)
plt.show()
```
